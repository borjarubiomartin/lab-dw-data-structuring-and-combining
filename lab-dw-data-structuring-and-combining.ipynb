## Challenge 1: Combining & Cleaning Data

In this challenge, we will be working with the customer data from an insurance company, as we did in the two previous labs. The data can be found here:

But this time, we got new data, which can be found in the following 2 CSV files located at the links below.

- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv

Note that you'll need to clean and format the new data.

Observation:

- Another option would be to read the clean file you saved in the previous lab, and just clean the two new files and concatenate the three clean datasets

```python
# Challenge 1: Combining & Cleaning Data
import pandas as pd
import numpy as np

# URLs de los datasets
url1 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv"
url2 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv"
url3 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv"

# Cargar los datasets
print("Cargando datasets...")
df1 = pd.read_csv(url1)
df2 = pd.read_csv(url2)
df3 = pd.read_csv(url3)

print(f"Dataset 1 shape: {df1.shape}")
print(f"Dataset 2 shape: {df2.shape}")
print(f"Dataset 3 shape: {df3.shape}")

# Examinar las primeras filas de cada dataset
print("\n--- Dataset 1 ---")
print(df1.head())
print("\n--- Dataset 2 ---")
print(df2.head())
print("\n--- Dataset 3 ---")
print(df3.head())
```

```python
# Función de limpieza de datos
def clean_data(df):
    """Función para limpiar los datos del seguro"""
    # Copiar el dataframe
    df_clean = df.copy()
    
    # Normalizar nombres de columnas
    df_clean.columns = df_clean.columns.str.lower().str.replace(' ', '_').str.replace('-', '_')
    
    # Limpiar valores de género
    if 'gender' in df_clean.columns:
        df_clean['gender'] = df_clean['gender'].str.lower().replace({'f': 'female', 'm': 'male'})
    
    # Limpiar state (normalizar mayúsculas/minúsculas)
    if 'state' in df_clean.columns:
        df_clean['state'] = df_clean['state'].str.title()
    
    # Limpiar education
    if 'education' in df_clean.columns:
        df_clean['education'] = df_clean['education'].str.title()
    
    # Limpiar customer_lifetime_value (remover símbolos de moneda si existen)
    if 'customer_lifetime_value' in df_clean.columns:
        df_clean['customer_lifetime_value'] = pd.to_numeric(
            df_clean['customer_lifetime_value'].astype(str).str.replace('$', '').str.replace(',', ''), 
            errors='coerce'
        )
    
    # Remover duplicados
    df_clean = df_clean.drop_duplicates()
    
    # Remover filas completamente vacías
    df_clean = df_clean.dropna(how='all')
    
    return df_clean

# Limpiar cada dataset
print("Limpiando datasets...")
df1_clean = clean_data(df1)
df2_clean = clean_data(df2)
df3_clean = clean_data(df3)

# Combinar los datasets
combined_df = pd.concat([df1_clean, df2_clean, df3_clean], ignore_index=True)

print(f"\nDataset combinado shape: {combined_df.shape}")
print(f"Duplicados encontrados: {combined_df.duplicated().sum()}")

# Remover duplicados finales si los hay
combined_df = combined_df.drop_duplicates()

print(f"Dataset final shape después de limpiar duplicados: {combined_df.shape}")
print("\nPrimeras filas del dataset combinado:")
print(combined_df.head())
print("\nInformación del dataset:")
print(combined_df.info())
```

# Challenge 2: Structuring Data

In this challenge, we will continue to work with customer data from an insurance company, but we will use a dataset with more columns, called marketing_customer_analysis.csv, which can be found at the following link:

```python
# Challenge 2: Structuring Data
# Cargar el dataset de marketing
marketing_url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv"
marketing_df = pd.read_csv(marketing_url)

print(f"Marketing dataset shape: {marketing_df.shape}")
print("\nColumnas del dataset:")
print(marketing_df.columns.tolist())
print("\nPrimeras filas:")
print(marketing_df.head())
print("\nInformación del dataset:")
print(marketing_df.info())
print("\nValores únicos en sales_channel:")
if 'sales_channel' in marketing_df.columns:
    print(marketing_df['sales_channel'].value_counts())
elif 'Sales Channel' in marketing_df.columns:
    print(marketing_df['Sales Channel'].value_counts())
```

1. You work at the marketing department and you want to know which sales channel brought the most sales in terms of total revenue. Using pivot, create a summary table showing the total revenue for each sales channel (branch, call center, web, and mail).
Round the total revenue to 2 decimal points.  Analyze the resulting table to draw insights.

2. Create a pivot table that shows the average customer lifetime value per gender and education level. Analyze the resulting table to draw insights.

## Bonus

You work at the customer service department and you want to know which months had the highest number of complaints by policy type category. Create a summary table showing the number of complaints by policy type and month.
Show it in a long format table.

*In data analysis, a long format table is a way of structuring data in which each observation or measurement is stored in a separate row of the table. The key characteristic of a long format table is that each column represents a single variable, and each row represents a single observation of that variable.*

*More information about long and wide format tables here: https://www.statology.org/long-vs-wide-data/*

```python
# Pregunta 1: Total revenue por sales channel usando pivot table
print("=== PREGUNTA 1: Revenue por Sales Channel ===")

# Identificar las columnas correctas (pueden tener nombres diferentes)
revenue_col = None
sales_channel_col = None

for col in marketing_df.columns:
    if 'total' in col.lower() and ('claim' in col.lower() or 'revenue' in col.lower()):
        revenue_col = col
    elif 'sales' in col.lower() and 'channel' in col.lower():
        sales_channel_col = col

print(f"Columna de revenue: {revenue_col}")
print(f"Columna de sales channel: {sales_channel_col}")

if revenue_col and sales_channel_col:
    # Crear pivot table para revenue por sales channel
    revenue_pivot = marketing_df.pivot_table(
        values=revenue_col,
        index=sales_channel_col,
        aggfunc='sum'
    ).round(2)
    
    print("\nTotal Revenue por Sales Channel:")
    print(revenue_pivot.sort_values(revenue_col, ascending=False))
    
    # Análisis
    print("\n--- ANÁLISIS ---")
    best_channel = revenue_pivot.idxmax()[0]
    worst_channel = revenue_pivot.idxmin()[0]
    print(f"Canal con mayor revenue: {best_channel}")
    print(f"Canal con menor revenue: {worst_channel}")
else:
    print("No se encontraron las columnas esperadas. Columnas disponibles:")
    print(marketing_df.columns.tolist())

print("\n" + "="*60)

# Pregunta 2: Average Customer Lifetime Value por gender y education
print("=== PREGUNTA 2: CLV promedio por Gender y Education ===")

# Identificar columnas
clv_col = None
gender_col = None
education_col = None

for col in marketing_df.columns:
    if 'customer' in col.lower() and 'lifetime' in col.lower():
        clv_col = col
    elif 'gender' in col.lower():
        gender_col = col
    elif 'education' in col.lower():
        education_col = col

print(f"Columna CLV: {clv_col}")
print(f"Columna Gender: {gender_col}")
print(f"Columna Education: {education_col}")

if clv_col and gender_col and education_col:
    # Crear pivot table para CLV promedio
    clv_pivot = marketing_df.pivot_table(
        values=clv_col,
        index=gender_col,
        columns=education_col,
        aggfunc='mean'
    ).round(2)
    
    print("\nCustomer Lifetime Value promedio por Gender y Education:")
    print(clv_pivot)
    
    # Análisis
    print("\n--- ANÁLISIS ---")
    print("CLV promedio por género:")
    print(marketing_df.groupby(gender_col)[clv_col].mean().round(2))
    print("\nCLV promedio por educación:")
    print(marketing_df.groupby(education_col)[clv_col].mean().round(2))
else:
    print("No se encontraron todas las columnas necesarias")

print("\n" + "="*60)

# BONUS: Complaints por policy type y month en formato long
print("=== BONUS: Complaints por Policy Type y Month (Long Format) ===")

# Buscar columnas relacionadas con complaints, policy type y month
complaint_col = None
policy_col = None
month_col = None

for col in marketing_df.columns:
    if 'complaint' in col.lower() or 'number_of_complaints' in col.lower():
        complaint_col = col
    elif 'policy' in col.lower() and 'type' in col.lower():
        policy_col = col
    elif 'month' in col.lower():
        month_col = col

print(f"Columna Complaints: {complaint_col}")
print(f"Columna Policy Type: {policy_col}")
print(f"Columna Month: {month_col}")

if complaint_col and policy_col and month_col:
    # Crear tabla de complaints por policy type y month
    complaints_summary = marketing_df.groupby([policy_col, month_col])[complaint_col].sum().reset_index()
    complaints_summary.columns = ['Policy_Type', 'Month', 'Total_Complaints']
    
    print("\nComplaints por Policy Type y Month (Long Format):")
    print(complaints_summary.sort_values('Total_Complaints', ascending=False))
    
    # Análisis adicional
    print("\n--- ANÁLISIS ---")
    print("Top 5 combinaciones con más complaints:")
    print(complaints_summary.nlargest(5, 'Total_Complaints'))
    
    print("\nComplaints por mes:")
    monthly_complaints = complaints_summary.groupby('Month')['Total_Complaints'].sum().sort_values(ascending=False)
    print(monthly_complaints)
    
    print("\nComplaints por tipo de póliza:")
    policy_complaints = complaints_summary.groupby('Policy_Type')['Total_Complaints'].sum().sort_values(ascending=False)
    print(policy_complaints)
else:
    print("No se encontraron todas las columnas necesarias para el análisis de complaints")
    print("Columnas disponibles:")
    for i, col in enumerate(marketing_df.columns):
        print(f"{i+1}. {col}")
```
